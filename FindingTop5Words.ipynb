```python
url = 'https://nasional.kompas.com/read/2023/05/01/12594161/kpu-catat-ada-700-bakal-calon-dpd-penuhi-syarat-terbanyak-dari-jawa-barat'

article = Article(url)
```

```python
teks1 = article.summary
```

```python
teks = article.text
```

```python
# Hapus hashtag
teks = re.sub(r'#\w+', '', teks)

# Hapus kata-kata https, www, dan .com
teks = re.sub(r'https?:\/\/\S+', '', teks)
teks = re.sub(r'www\.\S+', '', teks)
teks = re.sub(r'\b\w+\.com\b', '', teks)

# Hapus simbol
teks = re.sub(r'[^\w\s]', '', teks)

# Ubah seluruh teks menjadi lowercase
teks = teks.lower()
```

```python
# Tokenisasi
tokens = word_tokenize(teks)

# Hapus stopwords
stop_words = set(stopwords.words('indonesian'))
tokens = [token for token in tokens if token not in stop_words]
```

```python
# Hitung frekuensi kemunculan kata
word_counts = Counter(tokens)
```

```python
# Ambil kata-kata dengan frekuensi kemunculan minimal 2
selected_words = [word for word, count in word_counts.items() if count >= 2]

# Size vektor 50
selected_words = sorted(selected_words, key=lambda x: word_counts[x], reverse=True)[:50]

# Buat set vektor
vector_set = {}
for i, word in enumerate(selected_words):
    vector_set[word] = i
```

```python
# Hitung IDF
N = len(teks)
idf = defaultdict(float)
for word in word_counts.keys():
    count = sum(1 for teks in teks if word in teks)
    idf[word] = math.log(N / (count + 1))
```

```python
# Filter kamus IDF untuk hanya kata-kata dalam vektor set
selected_idf = {k: v for k, v in idf.items() if k in vector_set}
```

```python
# Hitung TF-IDF untuk setiap kata dalam dokumen
tf_idf = {}
for i, text in enumerate([teks]):
    # Hitung frekuensi kemunculan kata-kata dalam dokumen menggunakan vektor set
    doc_word_counts = defaultdict(int)
    for token in word_tokenize(text):
        if token in vector_set:
            doc_word_counts[token] += 1

    # Hitung TF-IDF untuk setiap kata dalam dokumen
    doc_tf_idf = {}
    for word, count in doc_word_counts.items():
        doc_tf_idf[word] = count * selected_idf[word]

    # Simpan hasil ke dalam kamus tf_idf
    tf_idf[f"Berita 1"] = doc_tf_idf
```

```python
# Konversi kamus tf_idf ke dalam DataFrame dengan index yang diambil dari selected_words
df = pd.DataFrame(tf_idf, index=selected_words)
df = df = df.sort_values(by='Berita 1', ascending=False)

# Tampilkan lima kata dengan nilai tertinggi
print(df.head(5))
```
